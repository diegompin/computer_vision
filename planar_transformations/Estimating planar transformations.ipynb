## Projective Transformation Model

The general form of the projective transformation model is given as follows:

\begin{equation}
\lambda \begin{bmatrix} x \\ y \\ 1
\end{bmatrix} 
= \begin{bmatrix} 
\phi_{11} & \phi_{12} & \phi_{13} \\
\phi_{21} & \phi_{22} & \phi_{23} \\
\phi_{31} & \phi_{32} & \phi_{33} \\
\end{bmatrix} \begin{bmatrix} u \\ v \\ 1\end{bmatrix} \enspace ,
\end{equation}

which in cartesian coordinates can be written as follows: 

\begin{equation}
x = \frac{\phi_{11}u + \phi_{12}v + \phi_{13}}{\phi_{31}u + \phi_{32}v + \phi_33} \\ 
y = \frac{\phi_{21}u + \phi_{22}v + \phi_{23}}{\phi_{31}u + \phi_{32}v + \phi_33} \enspace ,
\end{equation}

or shortly:

\begin{equation}
   \mathbf{x} = \mathbf{hom}[\mathbf{w}_i, \mathbf{\Phi}]
\end{equation}



## Learning Projective Parameters
The learning problem of the projective transformation model can be defined as 

\begin{equation}
\mathbf{\hat{\Phi}} = \operatorname*{arg\,min}_{\mathbf{\mathbf{\phi}}} \Bigg[ \sum_{i=1}^{I} (\mathbf{x}_i - \mathbf{hom}[\mathbf{w}_i, \mathbf{\Phi}])^T (\mathbf{x}_i - \mathbf{hom}[\mathbf{w}_i, \mathbf{\Phi}]) \Bigg] 
\end{equation}

\begin{equation}
= \operatorname*{arg\,min}_{\mathbf{\mathbf{\phi}}} \Bigg[ \sum_{i=1}^{I} \bigg( x_i - \frac{\phi_{11}u + \phi_{12}v + \phi_{13}}{\phi_{31}u + \phi_{32}v + \phi_33} \bigg)^2 + \bigg( y_i -  \frac{\phi_{21}u + \phi_{22}v + \phi_{23}}{\phi_{31}u + \phi_{32}v + \phi_33}\bigg)^2 \Bigg]
\end{equation}

The term $\mathbf{\Phi\tilde{w}}$ is

\begin{equation}
\mathbf{\Phi\tilde{w}} = \begin{bmatrix} 
\phi_{11}u & \phi_{12}v & \phi_{13} \\
\phi_{21}u & \phi_{22}v & \phi_{23} \\
\phi_{31}u & \phi_{32}v & \phi_{33} \\
\end{bmatrix}
\end{equation}

##Show that Equation 15.37 follows from Equation 15.36 [2].
Proceding with the following substitutions 

\begin{equation}
\phi_{11}u + \phi_{12}v + \phi_{13} = \phi_1 \\
\phi_{21}u + \phi_{22}v + \phi_{23} = \phi_2 \\
\phi_{31}u + \phi_{32}v + \phi_{33} = \phi_3 \\
\end{equation}

We have the following system of 3 equations $R_1$, $R_2$ and $R_1$ 
\begin{equation}
\begin{matrix}
& + 0 \phi_1 & - 1 \phi_2 & + y \phi_3 & = 0 \\
& + 1 \phi_1 & + 0 \phi_2 & - x \phi_3 & = 0 \\
& - y \phi_1 & + x \phi_2 & + 0 \phi_3 & = 0 \\
\end{matrix}
\end{equation}

Converting the system of equations aforementioned into an augmented matrix:

\begin{equation}
\left[
\begin{matrix}
0 & -1 & y \\
1 & 0 & -x \\
-y & x & 0 \\
\end{matrix}
\right |
\left.  
\begin{matrix}
0 \\
0 \\
0 \\
\end{matrix}
\right]
\end{equation}

Replacing the third row by with the sum of row 3, $x$ times row 1 and $y$ times row 2:

\begin{equation}
R_3 \leftarrow R_3 + x R_1 + yR_2 \\
\end{equation}
\begin{equation}
\left[
\begin{matrix}
0 & -1 & y \\
1 & 0 & -x \\
0 & 0 & 0 \\
\end{matrix}
\right |
\left.  
\begin{matrix}
0 \\
0 \\
0 \\
\end{matrix}
\right]
\end{equation}

Converting back to a system of equations:
\begin{equation}
\begin{matrix}
& + 0 \phi_1 & - 1 \phi_2 & + y \phi_3 & = 0 \\
& + 1 \phi_1 & + 0 \phi_2 & - x \phi_3 & = 0 \\
& +0 \phi_1 & + 0 \phi_2 & + 0 \phi_3 & = 0 \\
\end{matrix}
\end{equation}
It can be seen that the last equation is in the form $0=0$. This resulted from equation 3 being a linear combination of the other two equations.

Therefore, although three linear constraints on the elements of $\mathbf{\Phi}$ were provided, only the first two equations are independent and the third can be discarded. 

Then, stacking vertically the two first equations for each of the pair of the $I$ points ${\mathbf{x_i},\mathbf{w_i}}$ we can for the system

\begin{equation}
\begin{bmatrix}
0 & 0 & 0 & -u_1 & -v_1 & -1 & y_1u_1 & y_1v_1 & y_1 \\
u_1 & v_1 & 1 & 0 & 0 & 0 & -x_1u_1 & -x_1v_1 & -x_1 \\
0 & 0 & 0 & -u_2 & -v_2 & -1 & y_2u_1 & y_2v_2 &y_2 \\
u_2 & v_2 & 1 & 0 & 0 & 0 & -x_2u_2 & -x_2v_2 &-x_2 \\
\vdots & \vdots &\vdots &\vdots &\vdots &\vdots &\vdots &\vdots &\vdots \\
0 & 0 & 0 & -u_I & -v_I & -1 & y_Iu_I & y_Iv_I &y_I \\
u_I & v_I & 1 & 0 & 0 & 0 & -x_Iu_I & -x_Iv_I &-x_I \\
\end{bmatrix} 
\begin{bmatrix} 
\phi_{11} \\ \phi_{12} \\ \phi_{13} \\ 
\phi_{21} \\ \phi_{22} \\ \phi_{23} \\  
\phi_{31} \\ \phi_{32} \\ \phi_{33} \\  
\end{bmatrix}= \mathbf{0} \enspace ,
\end{equation}

which has the form $\mathbf{A\Phi} = \mathbf{0}$

## Homography

### Projective transformation model

```python
def calculate_homography(W, X):
    nbr_correspondences = W.shape[1]
    A = np.zeros((2*nbr_correspondences,9))
    for i in range(nbr_correspondences):        
        A[2*i] = [0,0,0,-W[0][i],-W[1][i],-1,
                    X[1][i]*W[0][i],X[1][i]*W[1][i],X[1][i]]
        A[2*i+1] = [-W[0][i],-X[1][i],-1,0,0,0,
                    X[0][i]*W[0][i],X[0][i]*W[1][i],X[0][i]]
    
    U,S,V = np.linalg.svd(A)
    H = V[8].reshape(3,3)
    H = np.linalg.inv(H)  
    H = H / H[2,2] 
    #U, S, V = np.linalg.svd(H)
    #H_hat = np.dot(U, V.T)    
    return H
```

```python
def warpfcn(x):
    return warpfc(x, H)
def warpfc(x, H):
    x = np.array([x[0],x[1],1]) 
    xt = np.dot(H,x)
    xt = xt/xt[2]
    return xt[0],xt[1]
```

### Nonlinear optimization cost funtion

```python
def cost(X, W, Phi):
    Phi = Phi.reshape(3,3)
    L = 0
    I = W.shape[1]
    for i in range(I):
        Xi = X[0:2,i].reshape((2,1))
        Wi = W[0:2,i].reshape((2,1))
        first_term = (Xi - warpfc(Wi, Phi)).T
        second_term = (Xi - warpfc(Wi, Phi))
        L += np.dot(first_term, second_term)
    return L
```

## Mosaicking Problem

### Correspondences points between images

```python
#x = ginput(4)
#top left, top right, bottom left, bottom right
delta = 700
xr2 = np.array([(147, 441), (377, 448), (375, 542), (144, 536)])
xr3 = np.array([(648, 433), (871, 428), (873, 520), (649, 522)])

xr3 += delta

xl4 = np.array([(840, 415), (950, 405), (955, 497), (843, 505)])
xl3 = np.array([(167, 431), (279, 429), (279, 522), (164, 523)])
xl3 += delta

# top left, top right, bottom right, bottom left
Xr2 = np.array([np.array([p[1],p[0],1]) for p in xr2]).T
Xr3 = np.array([np.array([p[1],p[0],1]) for p in xr3]).T

Xl3 = np.array([np.array([p[1],p[0],1]) for p in xl3]).T
Xl4 = np.array([np.array([p[1],p[0],1]) for p in xl4]).T
```

### Images and their respective correspondences points. These images are from Jan Erik Solem [2] book.

```python
imr = np.array(Image.open('images/Univ2.jpg'))
imc = np.array(Image.open('images/Univ3.jpg'))
iml = np.array(Image.open('images/Univ4.jpg'))

fig = plt.figure()
ax = fig.add_subplot(1,3,1)
ax.set_title('LEFT', size=20)
ax.imshow(iml)
ax.plot(Xl4[1], Xl4[0], 'bo')
ax.set_ylim(iml.shape[0], 0)
ax.set_xlim(0, iml.shape[1])
ax = fig.add_subplot(1,3,2)
ax.set_title('CENTER', size=20)
ax.plot(Xl3[1] - delta, Xl3[0] - delta, 'bo')
ax.plot(Xr3[1] - delta, Xr3[0] - delta, 'go')
ax.imshow(imc)
ax.set_ylim(imc.shape[0], 0)
ax.set_xlim(0, imc.shape[1])
ax = fig.add_subplot(1,3,3)
ax.set_title('RIGHT', size=20)
ax.plot(Xr2[1], Xr2[0], 'go')
ax.imshow(imr)    
ax.set_ylim(imr.shape[0], 0)
ax.set_xlim(0, imr.shape[1])
```

### Learning projective parameters

```python
H_23 = calculate_homography(Xr2, Xr3)
H_43 = calculate_homography(Xl4, Xl3)
```

### Improving found projective parameters using nonlinear optimization

```python
objective_function = (lambda phi: cost(Xr2, Xr3, phi))
result = op.minimize(fun=objective_function, x0=H_23.flatten(), method='Nelder-Mead')
H_23 = result.x.reshape(3,3)

objective_function = (lambda phi: cost(Xl4, Xl3, phi))
result = op.minimize(fun=objective_function, x0=H_43.flatten(), method='Nelder-Mead')
H_43 = result.x.reshape(3,3)
```

### Projecting left and right images

```python
im_g = np.zeros(shape=(imc.shape[0] + 2*delta,imc.shape[1]+2*delta,3))
im_gr = np.zeros(shape=im_g.shape)

for u in range(im_g.shape[0]):
    for v in range(im_g.shape[1]):
        x,y = warpfc([u,v], H_23)   
        x = int(x)
        y = int(y)
        if(0 <= x < imr.shape[0] and 0 <= y < imr.shape[1]):
            for col in range(3):
                im_gr[u,v,col] = imr[x,y, col]
                #im_g2[u,v,col] = np.subtract(255, im_g2[u,v,col])
                
im_gl = np.zeros(shape=im_g.shape)

for u in range(im_g.shape[0]):
    for v in range(im_g.shape[1]):
        x,y = warpfc([u,v], H_43)   
        x = int(x)
        y = int(y)
        if(0 <= x < iml.shape[0] and 0 <= y < iml.shape[1]):
            for col in range(3):
                im_gl[u,v,col] = iml[x,y, col]
                #im_g2[u,v,col] = np.subtract(255, im_g2[u,v,col])
  
```

### Overlapping images

```python
im_g[delta:delta+imc.shape[0], delta:delta+imc.shape[1],:] = imc
alpha = ((im_gr[:,:,0] * im_gr[:,:,1] * im_gr[:,:,2] ) > 0)
for col in range(3):
    im_g[:,:,col] = im_gr[:,:,col]*alpha + im_g[:,:,col]*(1-alpha)

alpha = ((im_gl[:,:,0] * im_gl[:,:,1] * im_gl[:,:,2] ) > 0)
for col in range(3):
    im_g[:,:,col] = im_gl[:,:,col]*alpha + im_g[:,:,col]*(1-alpha)

##TODO there is some problem with the colors    
plt.imshow(1 - im_g)
```

## Results and Problems
The result found by the linear algebra using homogeneous coordinates was good and the optimization process was not able to find a better solution. However, it can be seen that if we, for instance, change one of the values of the matrix, the non-linear optimization technique was able to reconstruct the matrix.

The used optimization method Nelder-Mead should be used mainly when the derivatives are unknown which is not the case. Then, a next step is to use a optimization technique which considers the gradient and the hessian such as the Newton-CG.

Besides, when the nonlinear optimization is not used the left and right images are shifted in the Mosaicking.
